# Exercise 4
Prove the sum rule, the product rule, and the chain rule

## Sum Rule
$$
\begin{align*}
\frac{d}{dx}(f(x) + g(x)) &= \lim_{h\rightarrow 0} \frac{f(x+h)+g(x+h) - \left(f(x) + g(x)\right)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)-f(x)+g(x+h)-g(x)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)-f(x)}{h}+ \lim_{h\rightarrow 0} \frac{g(x+h)-g(x)}{h}\\
&= \frac{df(x)}{dx}+ \frac{dg(x)}{dx}\\
\end{align*}
$$

## Product Rule
$$
\begin{align*}
\frac{d}{dx}(f(x)g(x)) &= \lim_{h\rightarrow 0} \frac{f(x+h)g(x+h) - f(x)g(x)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)g(x+h)- f(x+h)g(x) + f(x+h)g(x) - f(x)g(x)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)(g(x+h)- g(x)) + (f(x+h) - f(x))g(x)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)(g(x+h)- g(x))}{h} + \lim_{h\rightarrow 0}\frac{(f(x+h) - f(x))g(x)}{h}\\
&= \left(\lim_{h\rightarrow 0} f(x+h)\right) \left(\lim_{h\rightarrow 0}\frac{g(x+h)- g(x)}{h}\right) + \left(\lim_{h\rightarrow 0}\frac{(f(x+h) - f(x))}{h}\right)g(x)\\
&= f(x) \frac{dg(x)}{dx} + \frac{df(x)}{dx} g(x)\\
\end{align*}
$$

## Chain Rule
$$
\begin{align*}
\frac{d}{dx}(f(g(x)) &= \lim_{h\rightarrow 0} \frac{f(g(x+h)) - f(g(x))}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(g(x+h)) - f(g(x))}{g(x+h) - g(x)} \cdot \frac{g(x+h) - g(x)}{h}\\
&= \left(\lim_{h\rightarrow 0} \frac{f(g(x+h)) - f(g(x))}{g(x+h) - g(x)}\right) \cdot \left(\lim_{h\rightarrow 0} \frac{g(x+h) - g(x)}{h}\right)\\
&= \frac{df(g(x))}{dg(x)} \frac{dg(x)}{dx}\\
\end{align*}
$$

### Sine

$$
\begin{align*}
\frac{d\sin(t)}{dt} &= \lim_{\Delta t \rightarrow 0}\frac{\sin(t+\Delta t) - \sin(t)}{\Delta t}\\
&= \lim_{\Delta t \rightarrow 0}\frac{\sin(t)\cos(\Delta t) + \cos(t)\sin(\Delta t) - \sin(t)}{\Delta t}\\
&= \sin(t)\lim_{\Delta t \rightarrow 0}\frac{\cos(\Delta t) - 1}{\Delta t} + \cos(t)\lim_{\Delta t \rightarrow 0} \frac{\sin(\Delta t)}{\Delta t}\\
&= \sin(t)\cdot 0 + \cos(t)\cdot 1\\
&= \cos(t)\\
\end{align*}
$$

### Cosine

$$
\begin{align*}
\frac{d\cos(t)}{dt} &= \lim_{\Delta t \rightarrow 0}\frac{\cos(t+\Delta t) - \cos(t)}{\Delta t}\\
&= \lim_{\Delta t \rightarrow 0}\frac{\cos(t)\cos(\Delta t) - \sin(t)\sin(\Delta t) - \cos(t)}{\Delta t}\\
&= \cos(t)\lim_{\Delta t \rightarrow 0}\frac{\cos(\Delta t) - 1}{\Delta t} - \sin(t)\lim_{\Delta t \rightarrow 0} \frac{\sin(\Delta t)}{\Delta t}\\
&= \cos(t)\cdot 0 - \sin(t)\cdot 1\\
&= -\sin(t)\\
\end{align*}
$$

### Exponential

$$
\begin{align*}
\frac{de^t}{dt} &= \lim_{\Delta t \rightarrow 0}\frac{e^{(t+\Delta t)} - e^t}{\Delta t}\\
&= \lim_{\Delta t \rightarrow 0}\frac{e^te^{\Delta t} - e^t}{\Delta t}\\
&= e^t \lim_{\Delta t \rightarrow 0}\frac{e^{\Delta t} - 1}{\Delta t}\\
&= e^t \cdot 1\\
\end{align*}
$$

### Logarithm

$$
\begin{align*}
\frac{d \ln t}{dt} &= \lim_{\Delta t \rightarrow 0}\frac{\ln (t+\Delta t) - \ln t}{\Delta t}\\
&= \lim_{\Delta t \rightarrow 0}\frac{\ln (\frac{t+\Delta t}{t})}{\Delta t}\\
&= \lim_{\Delta t \rightarrow 0}\frac{\ln (1+\frac{\Delta t}{t})}{\Delta t}\\
&= \lim_{\Delta t \rightarrow 0} \ln \left(1+\frac{\Delta t}{t}\right)^{1/\Delta t}\\
&= \lim_{\Delta t \rightarrow 0} \ln \left(1+\frac{1/t}{1/\Delta t}\right)^{1/\Delta t}\\
&= \lim_{n \rightarrow \infty} \ln \left(1+\frac{1/t}{n}\right)^{n}\\
&= \ln e^{1/t}\\
&= 1/t
\end{align*}
$$

# Exercise 7

Show that position and velocity vectors are orthogonal (for a rotating particle)

### Solution

$$
\begin{align*}
r(t) &= 
\begin{bmatrix}
R\cos(\omega t)\\
R\sin(\omega t)
\end{bmatrix}\\
v(t) &= 
\begin{bmatrix}
-R\omega \sin(\omega t)\\
R\omega \cos(\omega t)\\
\end{bmatrix}\\
r(t) \cdot v(t) &= (R\cos(\omega t))(-R\omega \sin(\omega t)) + (R\sin(\omega t))(R\omega \cos(\omega t))\\
&= -R^2\omega \cos(\omega t)\sin(\omega t) + R^2\omega \sin(\omega t)\cos(\omega t)\\
&= 0\\
\end{align*}
$$

# Exercise 4
Finish evaluating $\int_0^{\pi/2} x \cos x\, \mathrm{d}x$

## Solution

Letting $u = x$ and $\mathrm{d}v = \cos x\, \mathrm{d}x$
$$
\begin{align*}
    \int_0^{\pi/2}x \cos x\, \mathrm{d}x &= \left. x \sin x\right|_0^{\pi/2} - \int_0^{\pi/2} \sin x\, dx\\
    &= \pi/2 - (-\cos \pi/2 + \cos 0)\\
    &= \pi/2 - 1\\
\end{align*}
$$
