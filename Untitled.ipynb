# Exercise 2
The derivative of the derivative is called the second derivative and is written $\frac{d^2 f(t)}{dt^2}$. Take the second derivative of each of the functions listed above.

### Solution

$$
\begin{align*}
f'(t) = 4t^3+9t^2-24t^2+1 \\
g'(x)=\cos(x)+\sin(x)\\
\theta'(\alpha) = e^\alpha + 1 + \ln \alpha\\
x'(t)= 2\sin t\cos t + \sin t\\
\end{align*}
$$

$$
\begin{align*}
f''(t) = 12t^2+18t-48t+1 \\
g''(x)=-\sin(x)+\cos(x)\\
\theta''(\alpha) = e^\alpha + 1 /\alpha\\
x''(t)= 2\cos (2t) + \cos t\\
\end{align*}
$$

```julia
x_vals = LinRange(-2, 3, 100)
y_vals = x_vals.^4 + 3 * x_vals.^3 - 12 * x_vals.^2 + x_vals .- 6
y_prime_vals = 4 * x_vals.^3 + 9 * x_vals.^2 - 24 * x_vals .+ 1
y_prime_prime_vals = 12 * x_vals.^2 + 18 * x_vals - 48 * x_vals

plot(x_vals, y_vals, label=L"f(t)");
plot(x_vals, y_prime_vals, label=L"df/dt");
plot(x_vals, y_prime_prime_vals, label=L"d^2f/dt^2");

grid("on");
legend();
```

```julia
x_vals = LinRange(-2*pi, 2*pi, 100)
y_vals = sin.(x_vals) - cos.(x_vals)
y_prime_vals = cos.(x_vals) + sin.(x_vals)
y_prime_prime_vals = -sin.(x_vals) + cos.(x_vals)

plot(x_vals, y_vals, label=L"g(x)");
plot(x_vals, y_prime_vals, label=L"dg/dx");
plot(x_vals, y_prime_prime_vals, label=L"d^2g/dx^2");

grid("on");
legend();
```

```julia
x_vals = LinRange(1, 2, 100)
y_vals = exp.(x_vals) + x_vals .* log.(x_vals)
y_prime_vals = exp.(x_vals) .+ 1 .+ log.(x_vals)
y_prime__prime_vals = exp.(x_vals) .+ 1 ./x_vals

plot(x_vals, y_vals, label=L"\theta(\alpha)");
plot(x_vals, y_prime_vals, label=L"d\theta/d\alpha");
plot(x_vals, y_prime__prime_vals, label=L"d^2\theta/d\alpha^2");

grid("on");
legend();
```

```julia
x_vals = LinRange(-2*pi, 2*pi, 100)
y_vals = sin.(x_vals).^2 - cos.(x_vals)
y_prime_vals = sin.(2 * x_vals) + sin.(x_vals)
y_prime_prime_vals = 2 * cos.(2 * x_vals) + cos.(x_vals)

plot(x_vals, y_vals, label=L"x(t)");
plot(x_vals, y_prime_vals, label=L"dx/dt");
plot(x_vals, y_prime_prime_vals, label=L"d^2x/dt^2");

grid("on");
legend();
```

# Exercise 3
Use the chain rule to find the derivatives of each of the following functions

$$
\begin{align*}
g(t) = \sin (t^2) - \cos (t^2)\\
\theta(\alpha) = e^{3\alpha} + 3\alpha \ln (3\alpha)\\
x(t)= \sin^2 (t^2) - \cos (t^2)\\
\end{align*}
$$

### Solution

$$
\begin{align*}
g'(t) = 2t\cos (t^2) + 2t\sin (t^2)\\
\theta'(\alpha) = 3e^{3\alpha} + 3 + 3 \ln (3\alpha)\\
x'(t)= 2(2t)\sin (t^2)\cos(t^2) + 2t\sin (t^2)\\
\end{align*}
$$

```julia
x_vals = LinRange(-2*pi, 2*pi, 1000)
y_vals = sin.(x_vals .^2) - cos.(x_vals.^2)
y_prime_vals = 2 * x_vals .* cos.(x_vals .^ 2) - 2* x_vals .* sin.(x_vals .^2)

plot(x_vals, y_vals, label=L"g(t)");
plot(x_vals, y_prime_vals, label=L"dg/dt");

grid("on");
legend();
```

```julia
x_vals = LinRange(1/3, 2, 1000)
y_vals = exp.(3 * x_vals) .+ 3 * x_vals .* log.(3 * x_vals)
y_prime_vals = 3 * exp.(3 * x_vals) .+ 3 .+ 3 * log.(3 * x_vals)

plot(x_vals, y_vals, label=L"\theta(\alpha)");
plot(x_vals, y_prime_vals, label=L"d\theta/d\alpha");

grid("on");
legend();
```

```julia
x_vals = LinRange(1/6, 3, 1000)
y_vals = sin.(x_vals .^2).^2 .- cos.(x_vals.^2)
y_prime_vals = 4 * x_vals .* sin.(x_vals.^2) .* cos.(x_vals.^2) .+ 2 * x_vals .* sin.(x_vals.^2)

plot(x_vals, y_vals, label=L"x(t)");
plot(x_vals, y_prime_vals, label=L"dx/dt");

grid("on");
legend();
```

# Exercise 4
Prove the sum rule, the product rule, and the chain rule

## Sum Rule
$$
\begin{align*}
\frac{d}{dx}(f(x) + g(x)) &= \lim_{h\rightarrow 0} \frac{f(x+h)+g(x+h) - \left(f(x) + g(x)\right)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)-f(x)+g(x+h)-g(x)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)-f(x)}{h}+ \lim_{h\rightarrow 0} \frac{g(x+h)-g(x)}{h}\\
&= \frac{df(x)}{dx}+ \frac{dg(x)}{dx}\\
\end{align*}
$$

## Product Rule
$$
\begin{align*}
\frac{d}{dx}(f(x)g(x)) &= \lim_{h\rightarrow 0} \frac{f(x+h)g(x+h) - f(x)g(x)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)g(x+h)- f(x+h)g(x) + f(x+h)g(x) - f(x)g(x)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)(g(x+h)- g(x)) + (f(x+h) - f(x))g(x)}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(x+h)(g(x+h)- g(x))}{h} + \lim_{h\rightarrow 0}\frac{(f(x+h) - f(x))g(x)}{h}\\
&= \left(\lim_{h\rightarrow 0} f(x+h)\right) \left(\lim_{h\rightarrow 0}\frac{g(x+h)- g(x)}{h}\right) + \left(\lim_{h\rightarrow 0}\frac{(f(x+h) - f(x))}{h}\right)g(x)\\
&= f(x) \frac{dg(x)}{dx} + \frac{df(x)}{dx} g(x)\\
\end{align*}
$$

## Chain Rule
$$
\begin{align*}
\frac{d}{dx}(f(g(x)) &= \lim_{h\rightarrow 0} \frac{f(g(x+h)) - f(g(x))}{h}\\
&= \lim_{h\rightarrow 0} \frac{f(g(x+h)) - f(g(x))}{g(x+h) - g(x)} \cdot \frac{g(x+h) - g(x)}{h}\\
&= \left(\lim_{h\rightarrow 0} \frac{f(g(x+h)) - f(g(x))}{g(x+h) - g(x)}\right) \cdot \left(\lim_{h\rightarrow 0} \frac{g(x+h) - g(x)}{h}\right)\\
&= \frac{df(g(x))}{dg(x)} \frac{dg(x)}{dx}\\
\end{align*}
$$
